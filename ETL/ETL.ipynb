```python
df_platform = pd.concat(plataformas, ignore_index= True) #Concatenamos los 4 dataframes de las diferentes plataformas en uno solo y facilitar el tratamiento de  los datos
```

```python
df_platform.drop(['show_id'], axis=1, inplace=True) #Eliminamos la columna 'show_id', debido a que no brinda informaci√≥n relevante
```

```python
df_platform.drop(['description'], axis=1, inplace=True) #Eliminamos la columna 'description', debido a que la informaci√≥n no es relevante para el ejercicio
```

```python
df_platform.drop(['director'], axis=1, inplace=True) #Eliminamos la columna 'director', debido a que la informaci√≥n no es relevante para el ejercicio
```

```python
df_platform.drop(['date_added'], axis=1, inplace=True) #Eliminamos la columna 'date_added', debido a que la informaci√≥n no es relevante para el ejercicio
```

```python
df_platform.drop(['rating'], axis=1, inplace=True) #Eliminamos la columna 'rating', debido a que la informaci√≥n no es relevante para el ejercicio
```

```python
df_platform.drop(['country'], axis=1, inplace=True) #Eliminamos la columna 'country', debido a que la informaci√≥n no es relevante para el ejercicio
```

```python
df_platform.head(-5)
```

```python
df_platform['duration'] = df_platform['duration'].str.replace('min','').str.replace('Seasons','').str.replace('Season','') #Quitamos el texto correspondiente a la duraci√≥n, entendiendo que 'Movie' esta en minutos y 'TV Show' en Seasons
df_platform.head(-5)
```

```python
df_platform['duration'] = pd.to_numeric(df_platform['duration']) #Se convierte la columna 'duration' a float
```

```python
df_platform['title'] = df_platform['title'].str.strip() #Se eliminan los espacion innecesarios al final o al inicio de los nombres
df_platform['cast'] = df_platform['cast'].str.strip()
df_platform['listed_in'] = df_platform['listed_in'].str.strip()
```
